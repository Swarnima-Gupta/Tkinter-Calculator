```python
from tkinter import*
import math
```

```python
def add():                                     #to add 2 numbers (e1+e2)
    res = float(e1.get()) + float(e2.get())
    myText.set(res)

def sub():                                      #to subtract 2 numbers (e1-e2)
    res=float(e1.get())-float(e2.get())
    myText.set(res)
    
def mul():                                     #to multiply 2 numbers (e1*e2)
    res=float(e1.get())*float(e2.get())
    myText.set(res) 
    
def div():                                    #to divide 2 numbers (e1/e2)
    res=float(e1.get())/float(e2.get())
    myText.set(res)
    
def xor():                                    #to get the xor result of 2 numbers (e1^e2)
    res=float(e1.get())^float(e2.get())
    myText.set(res)
    
def mod():                                   #to get mod result of 2 numbers (e1%e2)
    res=float(e1.get())%float(e2.get())
    myText.set(res)  
    
def factorial(n):                                          # to find factorial of 1st number (e1!)
    return 1 if (n==1 or n==0) else n * factorial(n - 1)
    #res= math.factorial(float(n))
    ##res=float(e1.get())**2
    ##myText.set(res)
    ##res = factorial(float(e1.get()))
    #myText.set(res)
    
def calculate():
    res=factorial(int(e1.get()))
    #info.config(text=res)
    myText.set(res)
    
def log():                                    #to find log of 1st number (log(e1))
    res= math.log(float(e1.get()))
    #res=float(e1.get())**2
    myText.set(res)   
    
def square():                                #to find square of 1st number (e1**2)
    res=float(e1.get())**2
    myText.set(res)   
    
def cube():                                   #to find cube of 1st number (e1**3)
    res=float(e1.get())**3
    myText.set(res) 
    
def sqroot():                               #to find square root of 1st number (e1**(1/2))
    res=float(e1.get())**(1/2)
    myText.set(res) 
    
def cbroot():                              #to find cube root of 1st number (e1**(1/3))
    res=float(e1.get())**(1/3)
    myText.set(res) 
    
def clear():                               #to clear the result block
    global res
    res=""
    myText.set("")
    
master = Tk()
myText = StringVar()
Label(master, text="Number 1").grid(row=0, sticky=W)
Label(master, text="Number 2").grid(row=1, sticky=W)
Label(master, text="Result:").grid(row=3, sticky=W)
result = Label(master, text="", textvariable=myText).grid(row=3, column=1, sticky=W)

master.configure(background="light blue")
master.title("Simple Calculator")

e1 = Entry(master)
e2 = Entry(master)

e1.grid(row=0, column=1)
e2.grid(row=1, column=1)

b = Button(master, text="ADD", command=add,bg="pink",fg="black",height=2, width=8)
b.grid(row=0, column=2, columnspan=2, rowspan=2, sticky=W + E + N + S, padx=5, pady=5)
b = Button(master, text="SUBTRACT", command=sub,bg="pink",fg="black",height=2,width=8)
b.grid(row=0, column=4, columnspan=2, rowspan=2, sticky=W + E + N + S, padx=5, pady=5)
b = Button(master, text="MULTIPLY", command=mul,bg="pink",fg="black",height=2,width=8)
b.grid(row=0, column=6, columnspan=2, rowspan=2, sticky=W + E + N + S, padx=5, pady=5)
b = Button(master, text="DIVIDE", command=div,bg="pink",fg="black",height=2,width=8)
b.grid(row=0, column=8, columnspan=2, rowspan=2, sticky=W + E + N + S, padx=5, pady=5)

b = Button(master, text="XOR", command=xor,bg="pink",fg="black",height=2,width=8)
b.grid(row=2, column=2, columnspan=2, rowspan=2, sticky=W + E + N + S, padx=5, pady=5)
b = Button(master, text="MODULUS", command=mod,bg="pink",fg="black",height=2,width=8)
b.grid(row=2, column=4, columnspan=2, rowspan=2, sticky=W + E + N + S, padx=5, pady=5)
b = Button(master, text="FACTORIAL", command=calculate,bg="pink",fg="black",height=2,width=8)
b.grid(row=2, column=6, columnspan=2, rowspan=2, sticky=W + E + N + S, padx=5, pady=5)
b = Button(master, text="LOG", command=log,bg="pink",fg="black",height=2,width=8)
b.grid(row=2, column=8, columnspan=2, rowspan=2, sticky=W + E + N + S, padx=5, pady=5)

b = Button(master, text="SQUARE", command=square,bg="pink",fg="black",height=2,width=8)
b.grid(row=4, column=2, columnspan=2, rowspan=2, sticky=W + E + N + S, padx=5, pady=5)
b = Button(master, text="CUBE", command=cube,bg="pink",fg="black",height=2,width=8)
b.grid(row=4, column=4, columnspan=2, rowspan=2, sticky=W + E + N + S, padx=5, pady=5)
b = Button(master, text="SQROOT", command=sqroot,bg="pink",fg="black",height=2,width=8)
b.grid(row=4, column=6, columnspan=2, rowspan=2, sticky=W + E + N + S, padx=5, pady=5)
b = Button(master, text="CBROOT", command=cbroot,bg="pink",fg="black",height=2,width=8)
b.grid(row=4, column=8, columnspan=2, rowspan=2, sticky=W + E + N + S, padx=5, pady=5)

b = Button(master, text="CLEAR(result)", command=clear,bg="pink",fg="black",height=2,width=10)
b.grid(row=10, column=6, columnspan=2, rowspan=2, sticky=W + E + N + S, padx=7, pady=7)

mainloop()
```
